```python
import requests
import json
import csv

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# URL do endpoint
url = "https://imunizacao-es.saude.gov.br/_search"
```

```python
# payload fazendo uma requisição de 10000 dados da tabela
payload = json.dumps({
  "size": 10000
})
# headers -cabeçalho- tem informações sobre autorização e tipo de conteúdo 
headers = {
  'Authorization': 'Basic aW11bml6YWNhb19wdWJsaWM6cWx0bzV0JjdyX0ArI1Rsc3RpZ2k=',
  'Content-Type': 'application/json'
}
```

```python
# response é a variável que receberá os dados pedidos na página 
response = requests.request("POST", url, headers=headers, data=payload)
```

```python
# imprimir os dados em texto
print(response.text)
```

```python
# variável que recebe os dados na linguagem JSON
data_json = response.json()
```

```python
# "filtra" os dados
data_json['hits']['hits']
```

```python
# normalizar os dados
df = pd.json_normalize(data_json['hits']['hits'])
```

```python
df.to_csv("dadosCovid.csv",index=False,sep=';',encoding="utf8")
```
